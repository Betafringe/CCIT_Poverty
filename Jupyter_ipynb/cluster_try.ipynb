# 使用 KNN 算法进行聚类，预测 status 相关性

```python
def load_traindata():    
    train_data = df.values[0:, 1:]
    train_label = df.values[:, 11]
    test_data = df.values[0:, 1:]
    return train_data, train_label, test_data
# def test_label_value():
#     test_label = df.values[1500:, 11]
#     return test_label
def knnClassify(trainData, trainLabel):
    knnClf = KNeighborsClassifier(n_neighbors=5)   # default:k = 5,defined by yourself:KNeighborsClassifier(n_neighbors=10)
    knnClf.fit(trainData, np.ravel(trainLabel)) # ravel Return a contiguous flattened array.
    return knnClf
def accuracy(raw, predict):
    raw_list = raw.tolist()
    predict_list = predict.tolist()
    count = 0
    for i in range(len(raw_list)):
        if raw_list[i] == predict_list[i]:
            count = count + 1
        else:
            pass
    print ("correct sample:",count, "all sample:", 
           len(raw_list), "accuracy rate: ", count/len(raw_list))
    
def main():
    init_time = time.time()
    trainData, trainLabel, test_data= load_traindata()
    print("data load finished!")
    knnClf = knnClassify(trainData, trainLabel)
    test_label_result = knnClf.predict(test_data)
    test_label_raw = df.values[1500:, 11]
    accuracy(test_label_raw, test_label_result)
#     for item in zip(test_label_result, test_label_raw):
#           print(item)
main()
```
