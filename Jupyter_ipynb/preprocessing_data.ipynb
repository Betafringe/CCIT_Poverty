```python
df_itter_income = df_drop_na[df_drop_na['income']< 100000]
df_itter_year = df_itter_income[df_itter_income['year']< datetime.datetime.now().year]
```

#更改 ID_CARD ，转换为年龄

```python
ID_card = df_itter_year['ID_card']
ID_list = []
for item in ID_card:
    sl_item = datetime.datetime.now().year - int(item[6:10])
    ID_list.append(sl_item)

df_itter_year.loc[:, 'ID_card'] = ID_list
```

```python
income_mean = df_itter_year['income'].mean()
df_itter_year['income'] = df_itter_year.loc[:,'income'].replace(0, income_mean)
```

```python
df_itter_year['out_poverty'] = df_itter_year['out_poverty'].replace('否','0')
df_itter_year['out_poverty'] = df_itter_year['out_poverty'].replace('是','1')
df_itter_year['level'] = df_itter_year['level'].replace('一般农户','0')
df_itter_year['level'] = df_itter_year['level'].replace('低保户','0')
df_itter_year['level'] = df_itter_year['level'].replace('低保贫苦户','0')
df_itter_year['level'] = df_itter_year['level'].replace('县级','1')
df_itter_year['level'] = df_itter_year['level'].replace('区级','1')
df_itter_year['level'] = df_itter_year['level'].replace('县','1')
df_itter_year['level'] = df_itter_year['level'].replace('县级示范','2')
df_itter_year['level'] = df_itter_year['level'].replace('省级','3')
df_itter_year['level'] = df_itter_year['level'].replace('省','3')
df_itter_year['level'] = df_itter_year['level'].replace('省级示范','3')
df_itter_year['level'] = df_itter_year['level'].replace('四川省','3')
df_itter_year['level'] = df_itter_year['level'].replace('国家级','4')
```
